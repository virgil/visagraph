```python
# -*- coding: utf-8 -*-

import json
from pprint import pprint

INPUT_FILENAME = 'visa_edgedata.json'
ABBREVIATIONS_FILENAME = 'country_abbrevs.json'
OUTPUT_FILENAME = 'new_visa_edgedata.json'



with open(INPUT_FILENAME,'r') as f:
    D = json.load(f)
    
#pprint(D)
```

```python
with open(ABBREVIATIONS_FILENAME,'r') as f:
    CCs = json.load(f)
    CCs = { x['Name']: x['Code'] for x in CCs }
    


CCs['Bolivia'] = 'BO'
CCs['British Virgin Islands'] = 'VG'
CCs['Brunei'] = 'BN'
CCs["Cote d'Ivoire"] = 'CI'
CCs['Democratic Republic of the Congo'] = 'CD'
CCs['East Timor'] = 'TL'
CCs['French West Indies'] = 'MQ'
CCs['Iran'] = 'IR'
CCs['Kosovo'] = 'XK'
CCs['Laos'] = 'LA'
CCs['Macau'] = 'MO'
CCs['Macedonia'] = 'MK'
CCs['Micronesia'] = 'FM'
CCs['Moldova'] = 'MD'
CCs['North Korea'] = 'KP'
CCs['Palestine'] = 'PS'
CCs["People's Republic of China"] = 'CN'
CCs["Republic of China (Taiwan)"] = 'TW'
CCs['Republic of the Congo'] = 'CG'
CCs['Reunion'] = 'RE'
CCs['Russia'] = 'RU'
CCs['South Korea'] = 'KR'
CCs['Syria'] = 'SY'
CCs['Tanzania'] = 'TZ'
CCs['United States Virgin Islands'] = 'VI'
CCs['Vatican City'] = 'VA'
CCs['Venezuela'] = 'VE'
CCs['Vietnam'] = 'VN'
#pprint(CCs)

all_ccs = sorted(list(set(CCs.values())))
for cc in all_ccs:
    this_country_name = [ x for x, this_cc in CCs.items() if this_cc == cc ][0]
    print(cc,"\t",this_country_name)
```

```python
z = {}
for from_node, to_node, edge_type in D:

    cc_from_node = CCs[ from_node ]
    cc_to_node = CCs[ to_node ]

    z.setdefault(cc_from_node,{})

    assert cc_to_node not in z[cc_from_node], "These should all be unique."
    
    
    #z[from_node][to_node] = edge_type
    z[ cc_from_node ][ cc_to_node ] = edge_type

print("Found", len(D), "edges.")

with open(OUTPUT_FILENAME,'w') as f:
    json.dump(z,f, indent=True, sort_keys=True, ensure_ascii=False)
    print("Wrote dictionary to file", OUTPUT_FILENAME)
```
